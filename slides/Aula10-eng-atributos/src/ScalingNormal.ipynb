```python
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# set seed for reproducibility
np.random.seed(0)
```

# Introdução

Muitas vezes, utilizamos o termo *scaling* e normalização de forma indistinta. No entanto, existem diferenças entre fazer o *scaling* dos dados e a normalização. *Scaling* utilizamos quando precisamos alterar a escala de uma variável. Ao colocar os números dentro de uma faixa de valores nós não alteramos a distribuição dos mesmos, mas garantimos que eles fiquem entre um valor mínimo e máximo. 

No entanto, o objetivo da normalização é transformar a distribuição dos dados em uma distribuição normal.

# Scaling

```python
# generate 1000 data points randomly drawn from an exponential distribution
original_data = np.random.exponential(size=1000)

# mix-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns=[0])

# plot both together to compare
fig, ax = plt.subplots(1,2)
sns.histplot(original_data, ax=ax[0], kde=True, stat="density", linewidth=0)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True, stat="density", linewidth=0)
ax[1].set_title("Scaled data")
```

# Normalization

```python
normalized_data = stats.boxcox(original_data)

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.histplot(original_data, ax=ax[0], kde=True, stat="density", linewidth=0)
ax[0].set_title("Original Data")
sns.histplot(normalized_data[0], ax=ax[1], kde=True, stat="density", linewidth=0)
ax[1].set_title("Normalized data")
```

# Log

Aplicar *log* sobre os dados não tem uma relação direta com *scaling* ou normalização. No entanto, é uma transformação que pode ser aplicada a um conjunto de dados, principalmente se este conjunto de dados tem um *range* muito pequeno. O objetivo da normalização é fazer com os valores da variável possam ficar mais "espalhadas" ao longo do eixo X. 

```python
log_data = np.log1p(original_data)

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.histplot(original_data, ax=ax[0], kde=True, stat="density", linewidth=0)
ax[0].set_title("Original Data")
sns.histplot(log_data, ax=ax[1], kde=True, stat="density", linewidth=0)
ax[1].set_title("Log data")
```
