# Geração dos dados

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
x = np.array([5,  7,   8, 11, 13, 15, 17, 20, 21, 25, 29, 32, 35, 37, 40, 42, 45, 47, 53, 55, 60, 65, 67, 70, 75, 80, 85, 92]).reshape((-1, 1))
y = np.array([50, 46, 39, 30, 22, 20, 18, 14, 10, 10, 15, 18, 22, 26, 28, 25, 22, 29, 35, 38, 45, 40, 42, 47, 50, 52, 55, 60])

plt.plot(x, y, 'ro')
plt.title('Dataset considerado. Quantidade de exemplos igual a '+ str(x.shape[0]))
```

## Qual é o erro aceitável?

# Separação em conjunto de treinamento e teste

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=1234)

#etapa opcional: apenas utilizada para facilitar o plot, onde os pontos sao ordenados
X_train, y_train = (list(t) for t in zip(*sorted(zip(X_train, y_train))))

plt.plot(X_train, y_train, 'ro')
```

# Criação de um modelo linear

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_train)
plt.plot(X_train, y_train, 'ro')   
plt.plot(X_train, y_pred, color='b')
plt.title('Modelo linear')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print('Resultados considerando o conjunto de treinamento')
print('RMSE = ', mean_squared_error(y_train, y_pred))
print('r2 = ', r2_score(y_train, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_train, y_pred)) 
print('\n')

y_pred = model.predict(X_test)
print('Resultados considerando o conjunto de teste')
print('RMSE = ', mean_squared_error(y_test, y_pred))
print('r2 = ', r2_score(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred)) 
```

# Criação de um modelo polinomial grau N

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

#transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer = PolynomialFeatures(2)
transformer.fit(X_train)
x_ = transformer.transform(X_train)

model = LinearRegression().fit(x_, y_train)
y_pred = model.predict(x_)
plt.plot(X_train, y_train, 'ro')   
plt.plot(X_train, y_pred, color='b')
plt.title('Modelo polinomial grau 2')
plt.show()
```

```python
print('Resultados considerando o conjunto de treinamento')
print('RMSE = ', mean_squared_error(y_train, y_pred))
print('r2 = ', r2_score(y_train, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_train, y_pred)) 
print('\n')

transformer.fit(X_test)
x_ = transformer.transform(X_test)
y_pred = model.predict(x_)
print('Resultados considerando o conjunto de teste')
print('RMSE = ', mean_squared_error(y_test, y_pred))
print('r2 = ', r2_score(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred)) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

transformer_1 = PolynomialFeatures(degree=8, include_bias=False)
transformer_1.fit(X_train)
x_1 = transformer_1.transform(X_train)

transformer_2 = FunctionTransformer(np.sin, validate=True)
transformer_2.fit(X_train)
x_2 = transformer_2.transform(X_train)

transformer_3 = FunctionTransformer(np.log1p, validate=True)
transformer_3.fit(X_train)
x_3 = transformer_3.transform(X_train)

x_ = np.concatenate((x_1, x_2, x_3), axis=1)

model = LinearRegression().fit(x_, y_train)
y_pred = model.predict(x_)
plt.plot(X_train, y_train, 'ro')   
plt.plot(X_train, y_pred, color='b')
plt.title('Modelo polinomial grau 8 + seno + log')
plt.show()
```

```python
print('Resultados considerando o conjunto de treinamento')
print('RMSE = ', mean_squared_error(y_train, y_pred))
print('r2 = ', r2_score(y_train, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_train, y_pred)) 
print('\n')

transformer_1.fit(X_test)
x_1 = transformer_1.transform(X_test)
transformer_2.fit(X_test)
x_2 = transformer_2.transform(X_test)
transformer_3.fit(X_test)
x_3 = transformer_3.transform(X_test)
x_ = np.concatenate((x_1, x_2, x_3), axis=1)

y_pred = model.predict(x_)
print('Resultados considerando o conjunto de teste')
print('RMSE = ', mean_squared_error(y_test, y_pred))
print('r2 = ', r2_score(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
```
