```python
import pandas as pd
# reading csv and removing useless columns
df = pd.read_csv("../../../dataset/breast-cancer-wisconsin.csv")
df = df.drop(columns=['id', 'Unnamed: 32'])
df.head()
```

# Data Analysis

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

# https://www.kaggle.com/uciml/breast-cancer-wisconsin-data
# no null values

sns.set_theme(style="dark")
sns.set_palette("colorblind")

sns.histplot(
    data=df,
    x="radius_mean", hue="diagnosis",
    multiple="stack",
    edgecolor=".3",
    linewidth=.5,
    log_scale=False,
)

# mean radius over 17.5 usually means malignant cancer
```

## Separating dataset into training and validation datasets
