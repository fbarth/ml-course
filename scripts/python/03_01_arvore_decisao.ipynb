# Árvores de decisão para problemas de classificação

## Exemplo de árvore de decisão sem poda

```python
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz

iris = load_iris()
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)

#imprimindo a arvore 
#dot_data = tree.export_graphviz(clf, out_file=None) 
dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True, rounded=True,  
                      special_characters=True)

graph = graphviz.Source(dot_data) 
graph
```

```python
predicted = clf.predict(iris.data)
```

```python
#
#Exemplo de possível overfitting do modelo
#
from sklearn.metrics import confusion_matrix
confusion_matrix(iris.target, predicted)
```

## Exemplo de árvore de decisão com poda

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(iris.data, iris.target)

#imprimindo a arvore 
#dot_data = tree.export_graphviz(clf, out_file=None) 
dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True, rounded=True,  
                      special_characters=True)

graph = graphviz.Source(dot_data) 
graph
```

```python
predicted = clf.predict(iris.data)
from sklearn.metrics import confusion_matrix
confusion_matrix(iris.target, predicted)
```

# Árvores de decisão para problemas de regressão

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
boston = load_boston()
regressor = DecisionTreeRegressor(random_state=0, max_depth=2)
model = regressor.fit(boston.data, boston.target)

dot_data = tree.export_graphviz(model, out_file=None, 
                      feature_names=boston.feature_names,  
                      class_names=boston.target,  
                      filled=True, rounded=True,  
                      special_characters=True)

graph = graphviz.Source(dot_data) 
graph
```
